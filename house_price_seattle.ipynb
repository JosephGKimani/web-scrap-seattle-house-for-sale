```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import matplotlib.pyplot as plt
```

```python
url='https://www.homesinseattlearea.com/search/quick/PDQVLZZZ3M4CMY1TK/Seattle-Homes-For-Sale/?src=google&campaign=272587458&content=604510865257&keyword=seattle%20house%20for%20sale&device=c&gad_source=1&gclid=CjwKCAiAq4KuBhA6EiwArMAw1Axp0JZwSkzmGcEmfTq54iPpEBLzdOaC2XmE2nXprE2ZnE-OlJHoWBoCqQgQAvD_BwE'
resp=requests.get(url)
soup=bs(resp.content,'html.parser')
```

```python
print(soup.prettify)
```

```python
address=soup.find_all('p', class_='full-address')
address_list=[]
price_list=[]
for add in address:
    add_h=add.text.strip()
    address_list.append(add_h)
prices=soup.find_all('div',class_='card--list-header pdq-old')


print(address_list)
```

```python
for price in prices:
    house_price=price.text.strip()
    price_list.append(house_price)
    

    
```

```python
print(price_list)
```

```python
df=pd.DataFrame({'Price':price_list})
```
